# Basic Plotting:

## line plot

```python


# Data points
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# Create a line plot
plt.plot(x, y)
# Customizing y-axis ticks
##plt.yticks([1, 4, 9, 16])
# Adding labels and a title
plt.xlabel('x-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')

# Display the plot|

plt.show()
```

## scatter plot

```python
# Data points
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# Create a scatter plot
plt.scatter(x, y)

# Adding labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Display the plot
plt.show()
```

## Bar Plots

```python


# Categories and corresponding values
categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]

# Create a bar plot
plt.bar(categories, values)

# Adding labels and a title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')

# Display the plot
plt.show()
```

## Pie plot

```python
import matplotlib.pyplot as plt

# Data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [15, 30, 45, 10]  # Values representing the proportions of each category

# pie chart
plt.figure(figsize=(6, 6))  # Optional: Adjust the figure size
plt.pie(sizes, labels=categories, autopct='%1.1f%%')
#you can add startangle=140 determines the initial angle at which the first slice starts (in degrees).
#plt.pie(sizes, labels=categories, autopct='%1.1f%%', startangle=140)

# Adding a title
plt.title('Pie Chart Example')

# Display the plot
plt.axis('equal')  # Ensures the pie chart is circular
plt.show()
```

## subplots

```python
import matplotlib.pyplot as plt
import numpy as np

#  sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x

# Create a figure with multiple subplots
plt.figure(figsize=(10, 4))  # Optional: Set the figure size

# Subplot 1: Line plot
plt.subplot(1, 3, 1)  # 1 row, 3 columns, and this is the 1st subplot
plt.plot(x, y1)
plt.title('Sin(x)')

# Subplot 2: Line plot
plt.subplot(1, 3, 2)  # 1 row, 3 columns, and this is the 2nd subplot
plt.plot(x, y2)
plt.title('Cos(x)')

# Subplot 3: Scatter plot
plt.subplot(1, 3, 3)  # 1 row, 3 columns, and this is the 3rd subplot
plt.scatter(x, y3)
plt.title('Scatter Plot')

# Adjust spacing between subplots
plt.tight_layout()

# Show the figure with subplots
plt.show()
```
